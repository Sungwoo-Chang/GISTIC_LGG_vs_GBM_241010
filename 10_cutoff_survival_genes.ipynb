```python
import pandas as pd
```

```python
mdel_pgain = pd.read_csv('survival_results/survival_mdel_pgain_tcga_glioma.csv', index_col = 0)
mdel_pgain
```

```python
# keep rows with p-value < 0.01
mdel_pgain = mdel_pgain[mdel_pgain['p_value'] < 0.01]

# keep rows with long_survival == low
mdel_pgain = mdel_pgain[mdel_pgain['long_survival'] == 'low']
```

```python
mdel_pgain_list = mdel_pgain.index.tolist()
```

```python
# save the list to a file
with open('survival_mdel_pgain_tcga_glioma_list.txt', 'w') as f:
    for item in mdel_pgain_list:
        f.write("%s\n" % item)
```
